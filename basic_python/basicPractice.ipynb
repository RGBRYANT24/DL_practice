- 实现最大公约数（gcd)最小公倍数的计算

```python
def gcd(x, y):
    (x, y) = (y, x) if x > y else (x, y)
    temp = x % y
    while temp != 0:
        (x, y) = (y, temp)
        temp = x % y
    return y

def lcm(x, y):
    return x * y //gcd(x, y)
```

```python
print(gcd(15, 9), '\t', lcm(15, 9))
```

#### 函数（参数）
- 可以修改传入的列表形参
- 可以通过切片法传入列表的拷贝，防止其对原来的列表进行修改
- 可以穿入不确定数目的参数

```python
def print_models(unprinted_designs, completed_models):
    while unprinted_designs:
        current_designs = unprinted_designs.pop()#弹出列表最后一个元素
        print(current_designs)
        completed_models.append(current_designs)

unprinted_designs = ['iphone', 'ipad', 'imac']
completed_models = []

if __name__ == '__main__':
    print_models(unprinted_designs[:], completed_models#)通过传入切片的方式，传入列表的拷贝，使函数不能修改原来的列表
    print()
    print(unprinted_designs)
    print_models(unprinted_designs, completed_models)#修改了传入的列表
    print()
    print(unprinted_designs)
```

```python
def build_profile(first, last, **user_info):
    profile = {}
    profile['first_name'] = first
    profile['last_name'] = last
    for key, value in user_info.items():
        profile[key] = value
    return profile

user_profile = build_profile('albert', 'einstein', location='princeton', field='physics')
print(user_profile)
```

##### yield函数
yield 的作用就是把一个函数变成一个 generator，带有 yield 的函数不再是一个普通函数，Python 解释器会将其视为一个 generator，调用 fab(5) 不会执行 fab 函数，而是返回一个 iterable 对象！在 for 循环执行时，每次循环都会执行 fab 函数内部的代码，执行到 yield b 时，fab 函数就返回一个迭代值，下次迭代时，代码从 yield b 的下一条语句继续执行，而函数的本地变量看起来和上次中断执行前是完全一样的，于是函数继续执行，直到再次遇到 yield。


```python
def fab(max):
    n, a, b = 0, 0, 1
    while n < max:
        yield b
        a, b = b, a + b
        n += 1

for val in fab(5):
    print(val, end='\t')
```

#### 集合
- 和数学上的集合一样，不允许有重复元素
- 可以进行交集、并集、差集运算

```python
set1 = {1, 2, 3, 3}
set2 = set(range(1, 10))
print(set1)
print(set2)
print('length: ', len(set1))

set4 = {num for num in range(1, 100) if num % 3 == 0 or num % 5 ==0}
print(set4)

#向集合中添加元素
set1.add(4)
print(set1)
set2.update([11, 12])
print(set2)

#删除元素
set2.discard(5)
print(set2)
set2.remove(1)
print(set2)
set2.pop()
print(set2)

#集合的交集、并集、差集、对称差运算
print(set1 & set2)#交集
print(set1.intersection(set2))

print(set1 | set2)#并集
print(set1.union(set2))

print(set1 - set2)#减法
print(set1.difference(set2))

print(set1 ^ set2)#去除公共元素的并集
print(set1.symmetric_difference(set2))

#判断子集和超集
print(set1 <= set2)#子集
print(set1.issubset(set2))

print(set2 >= set1)#超集
print(set2.issuperset(set1))
```

#### 字典
- key-value的键值对

```python
#创建字典的构造器语法
scores =  {'adr': 100, 'aaa': 90}
print(scores)

#字典的构造器语法
items1 = dict(one=1, two=2, three=3, four=4)
print(items1)
items2 = dict(zip(['a', 'b', 'c'], '123'))
print(items2)
items3 = {num : num ** 2 for num in range(1, 10)}
print(items3)

#通过key可以获得字典中对应的值
print(scores['adr'])

#对字典中所有的key-value对进行遍历
for key in scores:
    print(f'{key}: {scores[key]}')#格式化的输出方式
    print(key)#单个循环取出的就是key

#更新字典中的元素
scores['lsy'] = 101
scores.update(lsy2=102)
print(scores)

print(scores.get('lsy2'))#get方法获取key对应的值
print(scores)

#删除字典中的元素
print(scores.popitem())#删除了最后一个元素
print(scores)
print(scores.pop('adr'))#通过key删除指定的元素
print(scores)

#清空字典
scores.clear()
print(scores)
```

##### 继承
- 在括号内指定父类的名字
- 通过super()调用父类的方法

```python
class Battery():

    def __init__(self, battery_size=70):
        self.battery_size = battery_size
    
    def describe_battery(self):
        print('This car has a ' + str(self.battery_size) + ' -kWh battery')

class Car():

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading =  0
        self.gas_tank = 100

    def get_descriptive_name(self):
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name

    def read_odometer(self):
        print('This car has ' + str(self.odometer_reading) +' miles')

    def update_odometer(self, mileage):
        if self.odometer_reading < mileage:
            self.odometer_reading = mileage
        else:
            print('Can\'t roll back the odometer')
    
    def increment_odometer(self, miles):
        self.odometer_reading += miles

    def fill_gas_tank(self):
        self.gas_tank = 100

class ElectricCar(Car):

    def __init__(self, make, model, year):#初始化父类的属性
        super().__init__(make, model, year)
        self.battery_size = 70
        self.battery = Battery()#将一个实例作为属性

    def describe_battery(self):#定义子类的属性和方法
        print('This car has a ' + str(self.battery_size) + " -kWh battery")
    
    def fill_gas_tank(self):
        print('This car has no gas tank')


my_tesla = ElectricCar('tesla', 'modle s', '2016')
print(my_tesla.get_descriptive_name())

my_tesla.describe_battery()
my_tesla.fill_gas_tank()
my_tesla.battery.describe_battery()


        
```

通过try-except-else代码块进行处理
